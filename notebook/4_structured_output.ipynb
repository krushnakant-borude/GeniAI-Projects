```python
# its is in string format 

# -> pydantic helps you in python to make datatypes

from pydantic import BaseModel, Field

class RespoonseStructure(BaseModel):
    name:str=Field(description='Complete Name')
    email:str=Field(description='Email Adress')
    age:int=Field(description='Age')




# llm shoudl return the answeer in the form of obj of responsestructure


structured_llm=llm.with_structured_output(RespoonseStructure)

res1=structured_llm.invoke(f'please give me only name and email and age from the text {text}')

```

```python
res1.model_dump()
```

```python
from typing import List
class Movies(BaseModel):
    title:str=Field(description="Movie Title")
    year:int=Field(description="Movei release year")

class AllMovies(BaseModel):
    movie:List[Movies]

movie_llm=llm.with_structured_output(AllMovies)

res2=movie_llm.invoke('tell me about 5 sci-fi movie')
```
