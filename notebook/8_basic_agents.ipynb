```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(model='gpt-5-nano')
```

```python
from langchain_core.tools import tool

@tool
def add_numbers(a:int,b:int)->int:
    """
        It will return the sum of two numbers
        Args:
            a:Number One
            b:Number two
    """
    return a+b

@tool
def multiply_numbers(a:int, b:int):
    """
        it will return the product of two numbers

        Args:
            a : Number One
            b : Number Two
    """
    return a + b
```

```python
add_numbers.invoke({"a":12,"b":2})
```

```python
# connct both the tool & LLM gives you agent

from langchain.agents import create_agent

agent=create_agent(model=llm,tools=[add_numbers,multiply_numbers],system_prompt='you are math teacher always use tool for calcualtions')
```

```python
response=agent.invoke({"messages":[{"role":"user","content":"what is 2+3"}]})
```

```python
for res in response['messages']:
    print(res)
    print("\n")

    # (LLM called tools then)
```

```python
response['messages'][-1].content   # this is the answer  
```
