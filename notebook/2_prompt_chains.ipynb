```python
# dynamics proming chain of conversation

from IPython.display import display, Markdown

prompts=[
    {'role':'system','content':'you are a python developer'},
    {'role':'user','content':'difference between map & filter'}
]

res=llm.invoke(prompts)

print(repr(res.content))
```

PROMT TEMPLATES dyamic prompting

```python
model=ChatOpenAI(model='gpt-5-mini')
from langchain_core.prompts import ChatPromptTemplate

prompts=ChatPromptTemplate([
    {"role":'system','content':"you're a translator and translate input in {language}"},
    {"role":'user',"content":'{query}'}
])

prompts
```

```python
final_prompts=prompts.format_messages(language='hindi', query='I love my india, I love my india')

final_prompts
```

```python
res=model.invoke(final_prompts)

display(res.content)
```

## Chains-> Runnables

```python
res=model.invoke(prompts.invoke({'language':'hindi','query':'major subject is english'}))
display(res.content)
```

```python
# we can write a function
from langchain_core.output_parsers import StrOutputParser
out=StrOutputParser()

def uppercase(x:str):
    return x.upper()

chains=prompts | model | out | uppercase

chains
```

```python
# chain is kind of runnable that we


resp=chains.invoke({'language':'hinglish',"query":"I love langhchain"})
print(resp)
```
