```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(model='gpt-5-mini')
```

```python
# google serper -> free / google search -> paid
# create google serper account and get the api

from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain.agents import create_agent

search=GoogleSerperAPIWrapper()
```

```python
search.run('address of taj mahal')
```

```python
agent=create_agent(model=llm,tools=[search.run],system_prompt="you are a agent can search any question on internet ")

```

```python
question="who won the IPL 2026?"


response=agent.invoke({"messages":[{"role":"user","content":question}]})

response["messages"][-1].content
```
